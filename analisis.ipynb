```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from collections import Counter

# Descargar recursos de NLTK si es necesario
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")

# Leer el contenido del archivo de transcripción
transcription_file = "transcripciones/transcripcion1.txt"
with open(transcription_file, "r") as file:
    transcription = file.read()

# Tokenización y eliminación de palabras vacías
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

tokens = word_tokenize(transcription.lower())
cleaned_tokens = [lemmatizer.lemmatize(token) for token in tokens if token.isalnum() and token not in stop_words]

# Contar la frecuencia de las palabras
word_freq = Counter(cleaned_tokens)

# Obtener las palabras más comunes
most_common_words = word_freq.most_common(30)  
print("Palabras más comunes:", most_common_words)
```

```python
# Definir la ruta del archivo de transcripción
transcription_file = "transcripciones/transcripcion1.txt"

# Contador para almacenar la cantidad de palabras
word_count = 0

# Abrir el archivo y contar las palabras
with open(transcription_file, "r") as file:
    for line in file:
        # Dividir la línea en palabras y contarlas
        words = line.split()
        word_count += len(words)

# Imprimir la cantidad total de palabras
print(f"La cantidad total de palabras en el archivo {transcription_file} es: {word_count}")
```

```python
import spacy
from collections import Counter

# Cargar el modelo de spaCy en español
nlp = spacy.load("es_core_news_sm")

# Lista para almacenar los sustantivos de todas las partes de la transcripción
nouns = []

# Analizar cada parte de la transcripción por separado
for part in transc_parts:
    # Tokenizar el texto y etiquetar las partes del discurso
    doc = nlp(part)
    # Filtrar los sustantivos
    nouns.extend([token.text.lower() for token in doc if token.pos_ == "NOUN"])

# Contar la frecuencia de cada sustantivo
noun_freq = Counter(nouns)

# Obtener los 30 conceptos más repetidos
top_nouns = noun_freq.most_common(30)

# Imprimir los 30 conceptos más repetidos con su frecuencia
print("Conceptos más repetidos:")
for noun, freq in top_nouns:
    print(f"{noun}: {freq}")

```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import FreqDist, bigrams
from textblob import TextBlob
from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.sparse import dia_matrix


# Descargar los conjuntos de datos necesarios de NLTK
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

# Inicializar el lematizador
lemmatizer = WordNetLemmatizer()

# Definir las palabras vacías en español
stop_words = set(stopwords.words('spanish'))

# Lista de archivos de texto para analizar
files = ['transcripciones/transcripcion1.txt', 'transcripciones/transcripcion2.txt', 'transcripciones/transcripcion3.txt']

# Crear una lista para almacenar los resultados
results = []

for file_path in files:
    # Leer el archivo de texto
    with open(file_path, 'r') as file:
        text = file.read().replace('\n', '')

    # Convertir el texto a minúsculas
    text = text.lower()

    # Tokenizar el texto
    words = nltk.word_tokenize(text)

    # Eliminar las palabras vacías y lematizar las palabras restantes
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]

    # Unir las palabras de nuevo en un solo string
    text = ' '.join(words)

    # Crear un objeto TextBlob
    blob = TextBlob(text)

    # Obtener el análisis de sentimientos
    sentiment = blob.sentiment

    # Añadir los resultados a la lista
    results.append([sentiment.polarity, sentiment.subjectivity])

    # Análisis de frecuencia de palabras
    freq_dist = FreqDist(words)
    print(f'Frecuencia de palabras para {file_path}:')
    print(freq_dist.most_common(10))

    # Análisis de bigramas
    bigram_freq = FreqDist(list(bigrams(words)))
    print(f'Frecuencia de bigramas para {file_path}:')
    print(bigram_freq.most_common(10))

    # Análisis de co-ocurrencia de palabras
    count_model = CountVectorizer(ngram_range=(1,1)) # default unigram model
    X = count_model.fit_transform([text])
    Xc = (X.T * X) # this is co-occurrence matrix in sparse csr format
    Xc.setdiag(np.zeros(Xc.diagonal().shape)) # sometimes you want to fill same word cooccurence to 0
    print(f'Co-ocurrencias de palabras para {file_path}:')
    print(list(zip(count_model.get_feature_names_out(), np.asarray(Xc.sum(axis=0)).ravel())))


    # Análisis de la red de palabras
    G = nx.Graph()
    for bigram, freq in bigram_freq.most_common(10):
        word1, word2 = bigram
        G.add_edge(word1, word2, weight=freq)
    nx.draw(G, with_labels=True)
    plt.show()

# Crear un mapa de calor
plt.figure(figsize=(10, 6))
sns.heatmap(results, annot=True, cmap='coolwarm', xticklabels=['Polarity', 'Subjectivity'], yticklabels=['Reunión con empresas', 'Reunión con sector público', 'Reunión con la academia'])
plt.show()
```
